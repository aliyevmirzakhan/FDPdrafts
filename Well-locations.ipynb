```python
import numpy as np
import pandas as pd
import welly
import matplotlib
import matplotlib.pyplot as plt
```

```python
base_dir = 'tutorial/WellX-'
```

```python
X3 = welly.Well.from_las(base_dir+'3.las')
X4 = welly.Well.from_las(base_dir+'4.las')
X5 = welly.Well.from_las(base_dir+'5.las')
X6 = welly.Well.from_las(base_dir+'6.las')
```

```python
df3 = X3.df()
df4 = X4.df()
df5 = X5.df()
df6 = X6.df()
```

```python
df5.rename(columns = {"Y\tDEV": "Y DEV",
            "X\tDEV": "X DEV"}, inplace=True)
```

```python
df3['DEPTH'] = df3.index*3.28084
df4['DEPTH'] = df4.index*3.28084
df5['DEPTH'] = df5.index*3.28084
df6['DEPTH'] = df6.index*3.28084
```

```python
X3_xdata = np.array(df3['X DEV'])
X3_ydata = np.array(df3['Y DEV'])
X3_Zdata = np.array(df3['TVD'])

X4_xdata = np.array(df4['X DEV'])
X4_ydata = np.array(df4['Y DEV'])
X4_Zdata = np.array(df4['DEPTH'])

X5_xdata = np.array(df5['X DEV'])
X5_ydata = np.array(df5['Y DEV'])
X5_Zdata = np.array(df5['DEPTH'])

X6_xdata = np.array(df6['X DEV'])
X6_ydata = np.array(df6['Y DEV'])
X6_Zdata = np.array(df6['DEPTH'])

%matplotlib notebook

ax = plt.axes(projection = '3d')

ax.plot3D(X3_xdata, X3_ydata, X3_Zdata)
ax.plot3D(X4_xdata, X4_ydata, X4_Zdata)
ax.plot3D(X5_xdata, X5_ydata, X5_Zdata)
ax.plot3D(X6_xdata, X6_ydata, X6_Zdata)
ax.invert_zaxis()
```
