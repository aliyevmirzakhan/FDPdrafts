```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
basedir = 'Out/'
```

```python
well1 = pd.read_csv(basedir + "Well1-PorSw.csv")
well2 = pd.read_csv(basedir + "Well2-PorSw.csv")
well3 = pd.read_csv(basedir + "Well3-PorSw.csv")
well4 = pd.read_csv(basedir + "Well4-PorSw.csv")
well5 = pd.read_csv(basedir + "Well5-PorSw.csv")
well6 = pd.read_csv(basedir + "Well6-PorSw.csv")
```

```python
fig = plt.figure(figsize = (8,8))


ax1 = fig.add_subplot(231)
ax1.set_title("Well1 porosity")
plt.gca().hist(well1['density_Porosity'], edgecolor = 'black')

ax2 = fig.add_subplot(232)
ax2.set_title("Well2 porosity")
plt.gca().hist(well2['density_porosity'], edgecolor = 'black')

ax3 = fig.add_subplot(233)
ax3.set_title("Well3 porosity")
plt.gca().hist(well3['density_porosity'], edgecolor = 'black')

ax4 = fig.add_subplot(234)
ax4.set_title("Well4 porosity")
plt.gca().hist(well4['density_porosity'], edgecolor = 'black')

ax5 = fig.add_subplot(235)
ax5.set_title("Well5 porosity")
plt.gca().hist(well5['density_porosity'], edgecolor = 'black')

ax6 = fig.add_subplot(236)
ax6.set_title("Well6 porosity")
plt.gca().hist(well6['density_porosity'], edgecolor = 'black')

plt.subplots_adjust(wspace = 0.5)
```

```python
#Combining the porosity of Each well
por = np.concatenate([well1['density_Porosity'], well2['density_porosity'], well3['density_porosity'],
                      well4['density_porosity'], well5['density_porosity'], well6['density_porosity']])
plt.gca().set_title("Porosity General Histogram")
plt.hist(por, edgecolor = 'black', bins = 20)
```

```python
Sw = np.concatenate([well1['Sw'], well2['Sw'], well3['Sw'], well4['Sw'], well5['Sw'],  well6['Sw'],
                      ])
Sw = Sw[Sw < 1.0]
plt.hist(Sw, edgecolor = 'black', bins = 15)
plt.gca().set_title("Sw distribution")
```
