### Perm Correction to In-Situ stress and gas slippage

```python
#Helper Functions
def correct_perm2stress(k_amb, Pe, b):
    Pe = Pe * 0.00689476
    
    return k_amb*np.exp(-b*np.power(Pe, 0.5))

def correct_perm2klinkenberg(k_g, porosity, avg_p):
    
    b = 15.61*(k_g/porosity)**(-0.447)
    
    return k_g /(1 + b/avg_p)
```

### Fitting Function to corrected porosity and perm values

```python
from sklearn.linear_model import LinearRegression

x = np.array([poroperm_sand['porosity']]).reshape(-1,1)
y = np.log([poroperm_sand['Klinkenberg Perm']]).reshape(-1,1)

model = LinearRegression()

model.fit(x,y)

k = model.coef_
b = model.intercept_

r2 = model.score(x,y)

print("R2: {}".format(r2))
print("the slope: {}".format(k))
print("The intercept: {}".format(b))
```

```python
def compute_perm(k,b, phi):
    return np.exp(k*phi + b)
```

### Vshale Calculation

```python
def GR_Index(GR_data):
    mnGR = min(GR_data)
    mxGR = max(GR_data)
    
    return (GR_data - mnGR)/(mxGR - mnGR)
def calc_Vshale (Igr):
    return 0.33*(np.power(2,Igr)- 1)

GR = logDerived['GR']
Igr = GR_Index(GR)

print(min(GR), max(GR))
print(GR[0])
print(Igr[0])
Vshale = calc_Vshale(Igr)

print(Vshale)
logDerived['Vshale'] = Vshale
```
